# Bolinger Band Signal

```python
df = fdr.DataReader('KS11')
```

```python
# mav: 이평선
# volume: 거래량 표시
# kwargs = dict(type='candle',mav=(20),volume=False,figratio=(10,5),figscale=0.7)
kwargs = dict(type='candle',volume=False,figratio=(10,5),figscale=0.7)
```

```python
middle_band = df['Close'].rolling(window=20).mean()
upper_band = middle_band + df['Close'].rolling(window=20).std()
lower_band = middle_band - df['Close'].rolling(window=20).std()
```

```python
df['UpperB'] = upper_band
df['LowerB'] = lower_band
```

```python
# add bolinger band
apdict = mpf.make_addplot(df[['UpperB','LowerB']][-252:],width=0.5)
# apdict = mpf.make_addplot(df['UpperB'][-252:])

mpf.plot(df[-252:],**kwargs,style='charles',
         addplot = apdict)
```

```python
buy_signal = [np.nan]
for idx_loc in range(1, len(df)):
    close_today = df['Close'][idx_loc]
    lb_today = lower_band[idx_loc]
    clsoe_yesterday = df['Close'][idx_loc-1]
    lb_yesterday = lower_band[idx_loc-1]
    
    if clsoe_yesterday < lb_yesterday and close_today > lb_today: # 종가가 밴드하단 상향돌파
        buy_signal.append(df['Low'][idx_loc]) 
    else:
        buy_signal.append(np.nan)
        
buy_signal = pd.Series(buy_signal, index=df.index)
```

```python
# 표시할 기간
periods = 500

# add bolinger band
apdict = [mpf.make_addplot(df[['UpperB','LowerB']][-periods:],width=0.5),
          mpf.make_addplot(middle_band[-periods:],width=1),
          mpf.make_addplot(buy_signal[-periods:], type='scatter', markersize=100,marker='^', color='r')]


mpf.plot(df[-periods:],**kwargs,style='charles',
         addplot = apdict)
```
